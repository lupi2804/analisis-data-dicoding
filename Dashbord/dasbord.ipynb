```python
import pandas as pd
import streamlit as st
import seaborn as sns
import matplotlib.pyplot as plt

# Set style untuk visualisasi
sns.set(style='dark')

# Fungsi untuk membuat DataFrame rata-rata polutan
def create_pollutant_mean_df(df, pollutant):
    df_mean = df.groupby(by='station').agg({pollutant: 'mean'}).sort_values(by=pollutant, ascending=False).reset_index()
    df_mean.rename(columns={
        'station': 'Kota',
        pollutant: 'Rata-Rata'
    }, inplace=True)
    return df_mean

# Fungsi untuk membuat DataFrame polutan
def create_pollutant_df(df):
    pollutant_df = df[['TEMP', 'PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']]
    pollutant_df.rename(columns={
        'TEMP': 'Temperatur'
    }, inplace=True)
    return pollutant_df

# Fungsi untuk membuat DataFrame polutan per jam
def create_pollutant_hourly_df(df):
    pollutant_hourly_df = df.groupby(by='hour').agg({
        "PM2.5": 'mean',
        "PM10": 'mean',
        "SO2": 'mean',
        "NO2": 'mean',
        "CO": 'mean',
        "O3": 'mean'
    }).reset_index()
    return pollutant_hourly_df

# Membaca dataset
try:
    df = pd.read_csv('dataset.csv')
except FileNotFoundError:
    st.error("Dataset tidak ditemukan. Pastikan file 'dataset.csv' tersedia.")
    st.stop()
except pd.errors.EmptyDataError:
    st.error("Dataset kosong. Mohon periksa isinya.")
    st.stop()

# Konversi kolom datetime
datetime_columns = ["year", "month", "day", "hour"]  # Sesuaikan dengan dataset Anda
for column in datetime_columns:
    df[column] = pd.to_datetime(df[column])

# Mengubah kolom 'hour' menjadi integer
df['hour'] = df['hour'].dt.hour

# Membuat DataFrame untuk masing-masing polutan
pollutant_pm25 = create_pollutant_mean_df(df, 'PM2.5')
pollutant_pm10 = create_pollutant_mean_df(df, 'PM10')
pollutant_so2 = create_pollutant_mean_df(df, 'SO2')
pollutant_no2 = create_pollutant_mean_df(df, 'NO2')
pollutant_co = create_pollutant_mean_df(df, 'CO')
pollutant_o3 = create_pollutant_mean_df(df, 'O3')
pollutant_df = create_pollutant_df(df)
pollutant_hourly = create_pollutant_hourly_df(df)

# Dashboard Streamlit
st.title('Proyek Analisis Data: Air Quality')
st.caption('Silvi Dian Pramana | silvidianpramana@gmail.com')

st.header('Project')
st.write('Dataset yang digunakan untuk dashboard ini adalah dataset kualitas udara.')
st.write('Dataset kualitas udara ini memiliki 17 kolom.')
st.write(df.head(10))

st.header('Konsentrasi Polusi Paling Tinggi')
st.write('')
tab1, tab2, tab3, tab4, tab5, tab6 = st.tabs(['Polutan PM2.5', 'Polutan PM10', 'Polutan SO2', 'Polutan NO2', 'Polutan CO', 'Polutan O3'])
colors = ["#408ABF", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

with tab1:
    st.subheader('5 Kota dengan Konsentrasi PM2.5 Tertinggi')
    fig1, ax = plt.subplots(figsize=(18, 6))
    sns.barplot(x='Rata-Rata', y='Kota', data=pollutant_pm25.head(5), palette=colors)
    ax.set_xlabel('Rata-Rata Konsentrasi (µg/m³)', fontsize=15)
    ax.set_ylabel(None)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    st.pyplot(fig1)

with tab2:
    st.subheader('5 Kota dengan Konsentrasi PM10 Tertinggi')
    fig, ax = plt.subplots(figsize=(18, 6))
    sns.barplot(x='Rata-Rata', y='Kota', data=pollutant_pm10.head(5), palette=colors)
    ax.set_xlabel('Rata-Rata Konsentrasi (µg/m³)', fontsize=15)
    ax.set_ylabel(None)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    st.pyplot(fig)

with tab3:
    st.subheader('5 Kota dengan Konsentrasi SO2 Tertinggi')
    fig, ax = plt.subplots(figsize=(18, 6))
    sns.barplot(x='Rata-Rata', y='Kota', data=pollutant_so2.head(5), palette=colors)
    ax.set_xlabel('Rata-Rata Konsentrasi (µg/m³)', fontsize=15)
    ax.set_ylabel(None)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    st.pyplot(fig)

with tab4:
    st.subheader('5 Kota dengan Konsentrasi NO2 Tertinggi')
    fig, ax = plt.subplots(figsize=(18, 6))
    sns.barplot(x='Rata-Rata', y='Kota', data=pollutant_no2.head(5), palette=colors)
    ax.set_xlabel('Rata-Rata Konsentrasi (µg/m³)', fontsize=15)
    ax.set_ylabel(None)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    st.pyplot(fig)

with tab5:
    st.subheader('5 Kota dengan Konsentrasi CO Tertinggi')
    fig, ax = plt.subplots(figsize=(18, 6))
    sns.barplot(x='Rata-Rata', y='Kota', data=pollutant_co.head(5), palette=colors)
    ax.set_xlabel('Rata-Rata Konsentrasi (µg/m³)', fontsize=15)
    ax.set_ylabel(None)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    st.pyplot(fig)

with tab6:
    st.subheader('5 Kota dengan Konsentrasi O3 Tertinggi')
    fig, ax = plt.subplots(figsize=(18, 6))
    sns.barplot(x='Rata-Rata', y='Kota', data=pollutant_o3.head(5), palette=colors)
    ax.set_xlabel('Rata-Rata Konsentrasi (µg/m³)', fontsize=15)
    ax.set_ylabel(None)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    st.pyplot(fig)

st.header('Korelasi Polutan PM2.5 dengan Polutan Lain')
st.write('Dengan menggunakan visualisasi diagram _scatter plot_, dapat dilihat keterkaitan antara polutan PM2.5 dengan polutan lain. Keterkaitan kuat terlihat antara polutan PM2.5 dengan polutan PM10 dan CO. Keterkaitan lemah terlihat antara polutan PM2.5 dengan polutan SO2, NO2, dan O3.')
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(18, 18))

sns.scatterplot(x='PM2.5', y='PM2.5', data=pollutant_df, ax=ax[0][0], s=10)
ax[0][0].set_title("Polutan PM2.5", loc="center", fontsize=15)
ax[0][0].set_xlabel('PM2.5 (µg/m³)', fontsize=12)
ax[0][0].set_ylabel('PM2.5 (µg/m³)', fontsize=12)

sns.scatterplot(x='PM2.5', y='PM10', data=pollutant_df, ax=ax[0][1], s=10)
ax[0][1].set_title("Polutan PM10", loc="center", fontsize=15)
ax[0][1].set_xlabel('PM2.5 (µg/m³)', fontsize=12)
ax[0][1].set_ylabel('PM10 (µg/m³)', fontsize=12)

sns.scatterplot(x='PM2.5', y='SO2', data=pollutant_df, ax=ax[1][0], s=10)
ax[1][0].set_title("Polutan SO2", loc="center", fontsize=15)
ax[1][0].set_xlabel('PM2.5 (µg/m³)', fontsize=12)
ax[1][0].set_ylabel('SO2 (µg/m³)', fontsize=12)

sns.scatterplot(x='PM2.5', y='NO2', data=pollutant_df, ax=ax[1][1], s=10)
ax[1][1].set_title("Polutan NO2", loc="center", fontsize=15)
ax[1][1].set_xlabel('PM2.5 (µg/m³)', fontsize=12)
ax[1][1].set_ylabel('NO2 (µg/m³)', fontsize=12)

sns.scatterplot(x='PM2.5', y='CO', data=pollutant_df, ax=ax[2][0], s=10)
ax[2][0].set_title("Polutan CO", loc="center", fontsize=15)
ax[2][0].set_xlabel('PM2.5 (µg/m³)', fontsize=12)
ax[2][0].set_ylabel('CO (µg/m³)', fontsize=12)

sns.scatterplot(x='PM2.5', y='O3', data=pollutant_df, ax=ax[2][1], s=10)
ax[2][1].set_title("Polutan O3", loc="center", fontsize=15)
ax[2][1].set_xlabel('PM2.5 (µg/m³)', fontsize=12)
ax[2][1].set_ylabel('O3 (µg/m³)', fontsize=12)

st.pyplot(fig)

st.header('Rata-Rata Konsentrasi Polutan per Jam')
st.write('Dengan menggunakan visualisasi diagram garis, dapat dilihat pergerakan konsentrasi polutan masing-masing pada setiap jam. Dapat dilihat bahwa sebagian besar polutan (PM2.5, PM10, NO2, dan CO) memuncak pada waktu malam hari, polutan SO2 memuncak pada waktu siang hari, dan polutan O3 memuncak pada waktu sore hari.')
tab1, tab2, tab3, tab4, tab5, tab6 = st.tabs(['Polutan PM2.5', 'Polutan PM10', 'Polutan SO2', 'Polutan NO2', 'Polutan CO', 'Polutan O3'])

with tab1:
    st.subheader('Rata-Rata Konsentrasi PM2.5 per Jam')
    fig, ax = plt.subplots(figsize=(18, 6))
    ax.plot(
        pollutant_hourly['hour'],
        pollutant_hourly['PM2.5'],
        marker='o',
        linewidth=2
    )
    ax.set_xlabel('Jam', fontsize=15)
    ax.set_ylabel('Konsentrasi (µg/m³)', fontsize=15)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    plt.xticks(pollutant_hourly['hour'])
    st.pyplot(fig)

with tab2:
    st.subheader('Rata-Rata Konsentrasi PM10 per Jam')
    fig, ax = plt.subplots(figsize=(18, 6))
    ax.plot(
        pollutant_hourly['hour'],
        pollutant_hourly['PM10'],
        marker='o',
        linewidth=2
    )
    ax.set_xlabel('Jam', fontsize=15)
    ax.set_ylabel('Konsentrasi (µg/m³)', fontsize=15)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    plt.xticks(pollutant_hourly['hour'])
    st.pyplot(fig)

with tab3:
    st.subheader('Rata-Rata Konsentrasi SO2 per Jam')
    fig, ax = plt.subplots(figsize=(18, 6))
    ax.plot(
        pollutant_hourly['hour'],
        pollutant_hourly['SO2'],
        marker='o',
        linewidth=2
    )
    ax.set_xlabel('Jam', fontsize=15)
    ax.set_ylabel('Konsentrasi (µg/m³)', fontsize=15)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    plt.xticks(pollutant_hourly['hour'])
    st.pyplot(fig)

with tab4:
    st.subheader('Rata-Rata Konsentrasi NO2 per Jam')
    fig, ax = plt.subplots(figsize=(18, 6))
    ax.plot(
        pollutant_hourly['hour'],
        pollutant_hourly['NO2'],
        marker='o',
        linewidth=2
    )
    ax.set_xlabel('Jam', fontsize=15)
    ax.set_ylabel('Konsentrasi (µg/m³)', fontsize=15)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    plt.xticks(pollutant_hourly['hour'])
    st.pyplot(fig)

with tab5:
    st.subheader('Rata-Rata Konsentrasi CO per Jam')
    fig, ax = plt.subplots(figsize=(18, 6))
    ax.plot(
        pollutant_hourly['hour'],
        pollutant_hourly['CO'],
        marker='o',
        linewidth=2
    )
    ax.set_xlabel('Jam', fontsize=15)
    ax.set_ylabel('Konsentrasi (µg/m³)', fontsize=15)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    plt.xticks(pollutant_hourly['hour'])
    st.pyplot(fig)

with tab6:
    st.subheader('Rata-Rata Konsentrasi O3 per Jam')
    fig, ax = plt.subplots(figsize=(18, 6))
    ax.plot(
        pollutant_hourly['hour'],
        pollutant_hourly['O3'],
        marker='o',
        linewidth=2
    )
    ax.set_xlabel('Jam', fontsize=15)
    ax.set_ylabel('Konsentrasi (µg/m³)', fontsize=15)
    ax.tick_params(axis='y', labelsize=15)
    ax.tick_params(axis='x', labelsize=15)
    plt.xticks(pollutant_hourly['hour'])
    st.pyplot(fig)
```
